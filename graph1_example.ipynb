```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message: str

def compliment_node(state: AgentState) -> AgentState:
    """"Creates personalized compliments for people"""

    state['message'] = state['message'] + ", you're doing an amazing job leraning Langgrapgh!"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("complimenter", compliment_node)
graph.set_entry_point("complimenter")
graph.set_finish_point("complimenter")

app = graph.compile()
```

```python
result = app.invoke({"message": "Dave"})

result['message']
```
