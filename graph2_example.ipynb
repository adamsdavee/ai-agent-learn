```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
from functools import reduce
import operator
```

```python
class AgentState(TypedDict):
    name: str
    values: list[int]
    operation: str
    result: str
```

```python
def calculate_node(state: AgentState) -> AgentState:
    """"This function calulates the result of the list of the integers based on the type of operation"""

    print(state)

    if state['operation'] == "+" :
        answer = sum(state['values'])
    elif state['operation'] == "*" :
        answer = reduce(operator.mul, state['values'], 1)
    else:
        raise ValueError("Invalid Operation")
    

    state['result'] = f"Hi {state['name']}, your answer is: {answer}"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("calculator", calculate_node)
graph.set_entry_point("calculator")
graph.set_finish_point("calculator")

app = graph.compile()
```

```python
answer = app.invoke({"name": "Chukwudi David", "values": [1, 2, 3, 4], "operation": "*"})

print(answer['result'])
```
