```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    age: str
    skills: list[str]
    final: str
```

```python
def first_node(state: AgentState) -> AgentState:
    """"The first node in the sequence"""

    state['final'] = f"{state['name']}, welcome to the system! "
    return state

def second_node(state: AgentState) -> AgentState:
    """"The second node in the sequence"""

    state['final'] = state['final'] + f"You are {state['age']} years old! "
    return state

def third_node(state: AgentState) -> AgentState:
    """"The third node in the sequence"""

    result = ", ".join(state['skills'][:-1]) + " and " + state['skills'][-1]

    state['final'] = state['final'] + f"You have skills in: {result}"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("first_node", first_node)
graph.add_node("second_node", second_node)
graph.add_node("third_node", third_node)

graph.set_entry_point("first_node")
graph.add_edge("first_node", "second_node")
graph.add_edge("second_node", "third_node")
graph.set_finish_point("third_node")

app = graph.compile()
```

```python
result = app.invoke({"name": "David", "age": "24", "skills": ["JS", "TS", "Python"]})

print(result['final'])
```
